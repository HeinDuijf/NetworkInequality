# Peptic Ulcer

## Load

```python
with open('./empirical_networks/pud_final.pkl', 'rb') as f:
  G_pepticulcer = pickle.load(f)

n_agents = G_pepticulcer.number_of_nodes()
print(n_agents)

# Create a mapping from node names to indexes
mapping = {node: index for index, node in enumerate(G_pepticulcer.nodes())}

# Relabel the nodes in the graph
G_pepticulcer_indexed = nx.relabel_nodes(G_pepticulcer, mapping)
G = G_pepticulcer_indexed
```

## Plot

```python
plot_network_degree_distribution(G)
plot_loglog(G)
```

## Randomize, Stats, and Plot

## Randomize

```python
tot_edges = len(G.edges())
for p in [0,0.25,0.5,0.75,1]:
  frac_edges = int(p * tot_edges)
  randomized_network = randomize_network(G,n_edges=frac_edges)
  print(f'Randomization parameter p={p}')
  print('Randomization done, calculating statistics...')
  stats = network_statistics(randomized_network)
  for key in stats:
    print(f"{key}: {stats[key]}")
  plot_network_degree_distribution(randomized_network)
  plot_loglog(randomized_network)
  print('\n')
```

## Equalize

```python
tot_edges = len(G.edges())
for p in [0,0.25,0.5,0.75,1]:
  frac_edges = int(p * tot_edges)
  randomized_network = equalize(G,n =frac_edges)
  print(f'Randomization parameter p={p}')
  print('Randomization done, calculating statistics...')
  stats = network_statistics(randomized_network)
  for key in stats:
    print(f"{key}: {stats[key]}")
  plot_network_degree_distribution(randomized_network)
  plot_loglog(randomized_network)
  print('\n')
```
